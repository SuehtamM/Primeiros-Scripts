```python
#Jogo da Forca

from random import randint
from IPython.display import clear_output
```

```python
# Board (tabuleiro)
board = ['''

+---+
|   |
    |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
|   |
    |
    |
=========''', '''

 +---+
 |   |
 O   |
/|   |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/    |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/ \  |
     |
=========''']
```

```python
word_bank = ['casa', 'sala', 'boneco', 'parede', 'curso', 'pessoa', 'livro', 'caneta', 'computador', 'desenvolvedor',
        'administrador', 'empregado', 'bibliografia', 'perfume', 'pavimento', 'marceneiro', 'projetista', 'mantenedor',
            'carro', 'bicicleta', 'passarinho', 'cachorro', 'tribunal', 'vendaval']
```

```python
# Classe
class Hangman:
    
    # Método Construtor
    def __init__(self, word):
        self.word = list(word)
        self.hidden_letters = ["_" for letter in self.word]
        self.wrong_letters = []
        self.right_letters = []
        self.wrong_letters = set(self.wrong_letters)
        self.right_letters = set(self.right_letters)
        self.lives = 0
        print("Bem vindo(a)! (Nenhuma palavra possui ç ou letras com acento).")

    # Método para adivinhar a letra
    def guess(self, letter):
        self.letter = letter.lower()
        self.index = []
        self.index = set(self.index)
        if self.letter in self.word:
            
            count = 0
            for letter in self.word:
                if self.letter == letter:
                    self.index.add(count)
                    count += 1
                else:
                    count += 1
                    
            self.right_letters.add(self.letter)
            for i in self.index:
                self.hidden_letters[i] = self.letter
        
        else:
            self.wrong_letters.add(self.letter)
            self.lives +=1
        
        self.show_word()
            
    # Método para mostrar a palavra até o momento
    def show_word(self):
        for letter in self.hidden_letters:
            print(letter, end='')
        
    # Método para verificar se o jogo terminou
    def hangman_over(self):
        if self.lives == 6:
            game_on = False
            print("\n\nSem mais chances! Você perdeu!\nA palavra correta era:", end='')
            for i in self.word:
                print(i, end='')
            return game_on
        
        elif '_' not in self.hidden_letters:
            game_on = False
            print('\n\nParabéns! Você acertou todas as letras!')
            return game_on
        
    # Método para checar o status do game e imprimir o board na tela
    def print_game_status(self):
        print(board[self.lives])
        print('\n\nCertas:',set(self.right_letters))
        print('Erradas:',set(self.wrong_letters))
        print('Tentativas restantes:', 6 - self.lives)
```

```python
#Função que retorna a letra que o jogador escolher
def pick_letter():
    guessed_letter = input("Digite apenas 1 letra: ")
    
    while (len(guessed_letter)!= 1) or (guessed_letter not in ('qwertyuioplkjhgfdsazxcvbnm'or'QWERTYUIOPLKJHGFDSAZXCVBNM')):
        guessed_letter = input("Apenas 1 letra do alfabeto é permitida: ")
    
    return guessed_letter
```

```python
# Função para ler uma palavra de forma aleatória do banco de palavras
def rand_word():
    rn = randint(0,len(word_bank)-1)
    return word_bank[rn]
```

```python
#Função para jogar novamente
def again():
    ans = input('Deseja jogar novamente? (S ou N): ')
    while ans not in ['S','s','N','n']:
        ans = input("Digite apenas S ou N: ")
    if ans not in ['S','s']:
        return False
    else:
        return True
```

```python
#controle do loop de funcionamento do jogo
playing = True
while playing:
    word = rand_word()
    forca = Hangman(word)
    game_on = True
    while game_on:
        forca.print_game_status()
        forca.show_word()
        letter = pick_letter()
        forca.guess(letter)
        forca.print_game_status()
        game_over = forca.hangman_over()
        clear_output()
        if game_over == False:
            game_on = False
        else:
            clear_output()

    forca.print_game_status()
    forca.show_word()    
    forca.hangman_over()
    playing = again()
    clear_output()
```
